```python
import pandas as pd

sentiment = pd.read_csv("primetrade/sentiment.csv.csv")
trades = pd.read_csv("primetrade/trader_data.csv.csv")
```

```python
print(sentiment.head())
print(trades.head())
```

```python
import os

os.rename("primetrade/sentiment.csv.csv", "primetrade/sentiment.csv")
os.rename("primetrade/trader_data.csv.csv", "primetrade/trader_data.csv")
```

```python
sentiment = sentiment.drop_duplicates()
trades = trades.drop_duplicates()

# Handle missing values
trades = trades.dropna(subset=['Closed PnL'])
# Based on the head, there is no column 'leverage'. Skipping this line.
# trades['leverage'] = trades['leverage'].fillna(trades['leverage'].mean())

# Convert dates
sentiment['date'] = pd.to_datetime(sentiment['date'])
trades['Timestamp IST'] = pd.to_datetime(trades['Timestamp IST'], format="%d-%m-%Y %H:%M")

# Extract date
sentiment['Date'] = sentiment['date'].dt.date
trades['Date'] = trades['Timestamp IST'].dt.date
```

```python
merged = trades.merge(sentiment, on='Date', how='inner')

print(merged.head())
print(merged.columns)
```

```python
print(merged['classification'].value_counts())
```

```python
merged['win'] = merged['Closed PnL'] > 0
```

```python
daily_pnl = merged.groupby('Date')['Closed PnL'].sum()
```

```python
trades_per_day = merged.groupby('Date').size()
```
